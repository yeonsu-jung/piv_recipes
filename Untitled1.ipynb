```python
def merge_dicts(*dict_args):
    """
    Given any number of dictionaries, shallow copy and merge into a new dict,
    precedence goes to key-value pairs in latter dictionaries.
    """
    result = {}
    for dictionary in dict_args:
        result.update(dictionary)
    return result
```

```python
sd = {'path': 'black_1_1_1_10_[02-18]'}

small = {"winsize": 16, "searchsize": 20, "overlap": 8,}
middle = {"winsize": 32, "searchsize": 36, "overlap": 16}

scales = {"pixel_density": 43.8,"scale_factor": 1e4, "arrow_width": 0.001}
bounds = {"u_upper_bound": 300,"u_lower_bound": -300,"v_upper_bound": 300,"v_lower_bound": -2000}
tr = {"transpose": False}
ic = {"show_result": True}
crop = {"crop": [0,0,0,0]}

piv_cond = merge_dicts(small,
                       scales,bounds,tr,ic,crop)
pi.set_piv_param(piv_cond)


pi.quick_piv(sd,index_a = 101,index_b = 102)
```
