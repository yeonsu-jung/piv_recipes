```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
N = 5
x = np.linspace(0,1,N)
y = np.linspace(0,1,N)

X,Y = np.meshgrid(x,y)

u = np.ones((N,N))
v = np.ones((N,N))
```

```python
fig, ax = plt.subplots(2)
ax[0].quiver(X,Y,u,v,angles='uv')
ax[1].quiver(X,Y,u,v,angles='xy')
```
